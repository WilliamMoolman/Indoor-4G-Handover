```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)

# ax.set_xlim([-0.05, 1])
# ax.set_ylim([-0.05, 1])
ax.set_xlabel('Time')
ax.set_ylabel('RSRP')
ax.set_title('Handover Example')

# Remove ticks
ax.set_xticks([])
ax.set_yticks([])

# Change axis to arrows
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['top'].set_color('none')

# Add arrows
# ax.annotate('', xy=(0.5, 0), xytext=(0.5, 2),
#             arrowprops=dict(arrowstyle='<->'))
ax.arrow(0, 0, 1, 0, head_width=0.02, head_length=0.03, length_includes_head=True,fc='k', ec='k')
ax.arrow(0, 0, 0, 1, head_width=0.02, head_length=0.03, length_includes_head=True,fc='k', ec='k')

# Add two curvy signals
# ax.annotate('', xy=(0.5, 0), xytext=(0.5, 0.5),
#             arrowprops=dict(arrowstyle='<->'))
xs = np.arange(0, 1, 0.01)
ys = 0.01*np.sin(xs * np.pi * 15 + 0.5) + 0.01*np.cos(xs * np.pi * 7) + xs*0.9
ax.plot(xs, ys, 'k', linestyle='--', label='Target Cell')

xs2 = np.arange(0, 1, 0.01)
ys2 = 0.02*np.sin(xs2 * np.pi * 6) + 0.04*np.cos(xs2 * np.pi * 13) - xs2*0.7 + +0.8
ax.plot(xs2, ys2, 'k', label='Source Cell')
ax.legend()

XVAL =70
# ax.arrow(xs[XVAL], ys[XVAL], 0, ys2[XVAL]-ys[XVAL], head_width=0.02, head_length=0.03, length_includes_head=True,fc='k', ec='k', arrowstyle='<->')
ax.annotate('', xy=(XVAL/100, ys[XVAL]), xytext=(XVAL/100, ys2[XVAL]),
            arrowprops=dict(arrowstyle='<->'))
ax.annotate('Hysteresis', xy=(XVAL/100+ 0.02, (ys2[XVAL]+ys[XVAL])/2-0.08), rotation=90)#, xytext=(XVAL/100, (ys2[XVAL]+ys[XVAL])/2))

ax.vlines(XVAL/100+0.2, 0, ys[XVAL+20], linestyle='dotted', color='k')

ax.annotate('Handover', xy=(XVAL/100+ 0.22, ys[XVAL+20]/2), rotation=90)

ax.annotate('', xy=(XVAL/100, 0.1), xytext=(XVAL/100+0.21, 0.1),
            arrowprops=dict(arrowstyle='|-|'))
ax.annotate('', xy=(XVAL/100, 0.1), xytext=(XVAL/100+0.21, 0.1),
            arrowprops=dict(arrowstyle='<->'))
ax.annotate('Time to Trigger', xy=(.4, 0.09))
```
