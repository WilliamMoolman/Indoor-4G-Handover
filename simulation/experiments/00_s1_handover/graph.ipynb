```python
import pandas as pd

df = pd.read_csv("rsrp_s1_handover.csv")
df
```

```python
# ENB is set to -68dBm with a multiply constant of 0.27
# Handover occured at 0.38 multiplier
import matplotlib.pyplot as plt

# Plot ENB2 with dotted linechart
plt.plot(df["ENB2_Gain"], df["ENB2_RSRP"], label="ENB2", marker="o")
plt.axhline(y=-68, color="r", linestyle="--", label="ENB1")
plt.axvline(x=0.38, color="g", linestyle="--", label="Handover")
plt.xlabel("Neighbour Gain")
plt.xlim(0.15, 0.45)
plt.ylim(-75, -63)
plt.ylabel("RSRP")
plt.legend()
plt.title("S1 Handover")
```
