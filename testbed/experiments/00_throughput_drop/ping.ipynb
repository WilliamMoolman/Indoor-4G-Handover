```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

pingfile = "data/ping-handover.csv"

df = pd.read_csv(pingfile)

df = df[df["Ping Response"] != 0]

plt.figure(figsize=(20, 10))

# fontsize
plt.rc("font", size=20)
plt.plot(df["Interval start"], df["Ping Response"], color="black", linewidth=1)


plt.xlabel("Time (s)")
plt.ylabel("Ping Response (ms)")

# grid
plt.grid(color="gray", linestyle="--", linewidth=1)
```
